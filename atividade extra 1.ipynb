```python
import math

# QUANTIZAÇÃO
def quantizacao():
    niveis = int(input("Níveis: "))
    bits = math.ceil(math.log2(niveis))  # calcula bits necessários

    print(f"Bits: {bits}")
    print(f"Combinações: {2**bits}")

    for i in range(2**bits):
        print(bin(i)[2:].zfill(bits))  # mostra em binário


# DECIMAL -> BINÁRIO
def decimal_para_binario():
    n = int(input("Decimal: "))
    print(bin(n)[2:])  # converte para binário


# DECIMAL COM FRAÇÃO
def decimal_fracao():
    n = float(input("Número: "))

    inteira = int(n)
    frac = n - inteira

    bin_int = bin(inteira)[2:]
    bin_frac = ""

    for _ in range(10):
        frac *= 2
        bit = int(frac)
        bin_frac += str(bit)
        frac -= bit
        if frac == 0:
            break

    print(f"{bin_int}.{bin_frac}")


# CONVERSOR DE BASES
def conversor():
    op = input("1-HEX | 2-OCT: ")

    if op == "1":
        print(int(input("Hex: "), 16))
    elif op == "2":
        print(int(input("Octal: "), 8))


# MENU
while True:
    print("\n1-Quantização  2-Binário  3-Fração  4-Bases  0-Sair")
    op = input("Opção: ")

    if op == "1":
        quantizacao()
    elif op == "2":
        decimal_para_binario()
    elif op == "3":
        decimal_fracao()
    elif op == "4":
        conversor()
    elif op == "0":
        break
```
