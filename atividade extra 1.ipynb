Atividade 1

```python
import math


def calcular_quantizacao():

    quantidade = int(input("Informe a quantidade de níveis: "))

    total_bits = math.ceil(math.log2(quantidade))


    print(f"\n--- Saída ---")

    print(f"Níveis informados: {quantidade}")

    print(f"Bits calculados: {total_bits}")

    print(f"Total possível com esses bits: {2**total_bits} combinações")


    print(f"\nCombinações com {total_bits} bits:")


    for valor in range(2**total_bits):

        codigo = f"{valor:0{total_bits}b}"

        print(f"Combinação {valor:3}: {codigo}")
calcular_quantizacao()
```

```python
def converter_decimal_binario(valor):

    if valor == 0:

        return "0"


    resultado = ""



    while valor > 0:

        resto = valor % 2

        resultado = str(resto) + resultado

        valor = valor // 2



    return resultado
n = int(input())

converter_decimal_binario(n)
```

```python
def dec_para_bin_fracao(numero, limite=10):

    # -------- PARTE INTEIRA --------
    parte_int = int(numero)
    valor = parte_int

    if valor == 0:
        bin_int = "0"
    else:
        resultado = ""
        while valor > 0:
            resto = valor % 2
            resultado = str(resto) + resultado
            valor = valor // 2
        bin_int = resultado

    # -------- PARTE FRACIONÁRIA --------
    parte_frac = numero - parte_int
    lista_bits = []

    contador = 0
    while parte_frac != 0 and contador < limite:
        parte_frac *= 2
        digito = int(parte_frac)
        lista_bits.append(str(digito))
        parte_frac -= digito
        contador += 1

    # -------- RESULTADO FINAL --------
    if lista_bits:
        return f"{bin_int}.{''.join(lista_bits)}"
    else:
        return bin_int


# teste
print(dec_para_bin_fracao(12.25))
```

```python

def hexa_para_bin(valor_hex):

    return bin(int(valor_hex, 16))[2:]



def octal_para_bin(valor_oct):

    return bin(int(valor_oct, 8))[2:]



def bin_para_decimal(texto_bin):

    soma = 0

    invertido = texto_bin[::-1]


    for indice, bit in enumerate(invertido):

        if bit == "1":

            soma += 2**indice

    return soma
n=input()
hexa_para_bin(n)
```
